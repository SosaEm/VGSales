```python
import requests
import json
import pandas as pd
from credenciales import credenciales
```

```python
url='https://api.rawg.io/api/games?key='+credenciales.get("SECRET_KEY")+ '&page_size=100'
r=requests.get(url)
print(r)
```

```python
games=r.json()
info=games["results"]
df = pd.json_normalize(info)
df.head()
```

```python
df.columns
```

```python
df1 = df.drop(['tba', 'ratings', 'added', 'platforms', 'tags', 'parent_platforms', 'added_by_status.yet', 'added_by_status.owned', 'background_image', 'added_by_status.dropped', 'added_by_status.playing',  'added_by_status.beaten', 'added_by_status.toplay', 'saturated_color', 'stores', 'dominant_color', 'clip', 'short_screenshots'], axis=1)
df1.columns
```

```python
df1.head()
```

**Voy a intentar "desempaquetar" la columna de generos**  //  Todavia no descubri como

```python
df_s = pd.DataFrame(pd.json_normalize(df1.genres, sep='_'))
df_s.head()
```

```python
df_s1 = pd.DataFrame(pd.json_normalize(df_s[0], sep='_'))
df_s1.drop(['id', 'slug', 'games_count', 'image_background'], axis=1, inplace=True)
df_s1.rename(columns={'name':'genre1'},inplace=True)
df_s2 = pd.DataFrame(pd.json_normalize(df_s[1], sep='_'))
df_s2.drop(['id', 'slug', 'games_count', 'image_background'], axis=1, inplace=True)
df_s2.rename(columns={'name':'genre2'},inplace=True)
df_s3 = pd.DataFrame(pd.json_normalize(df_s[2], sep='_'))
df_s3.drop(['id', 'slug', 'games_count', 'image_background'], axis=1, inplace=True)
df_s3.rename(columns={'name':'genre3'},inplace=True)
df_s4 = pd.DataFrame(pd.json_normalize(df_s[3], sep='_'))
df_s4.drop(['id', 'slug', 'games_count', 'image_background'], axis=1, inplace=True)
df_s4.rename(columns={"name":"genre4"},inplace=True)
```

```python
df1.drop('genres', axis=1, inplace=True)
df1.head()
```

```python
dfinal= pd.concat([df1, df_s1, df_s2, df_s3, df_s4], axis=1)
dfinal
```
